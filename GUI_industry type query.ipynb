```python
store = pd.read_csv('CAstore.csv')
store = store.sort_values(by='기준_년_코드', axis=0, ascending=True)
store
```

```python
#2021-2분기 store 데이터셋
latest_df = store[store['기준_년_코드'] == '2021_2']
latest_df
```

```python
#버튼 생성 클래스
class produce_buttons(Tk):
    def __init__ (self, root, data, colnum):
        self.root = root
        self.buttons = {}
        self.data = data
        self.colnum = colnum
        self.init_widgets()
    
    def init_widgets(self):
        for i,d in enumerate(self.data):
            global btn
            btn = Button(self.root)
            btn.config(text=f"{d}")
            btn.bind("<Button-1>", to_pop)
            btn.grid(row=2+i, column=self.colnum, padx=20, pady=5)
            self.buttons[i] = btn
            
def to_pop(event):
    global dist
    dist = btn.cget("text")
    pop_page()
        
def pop_page():
    pop_pg = Tk()
    pop_pg.geometry("800x600")
    pop_pg.option_add("*Font", "맑은고딕25")
    pop_pg.mainloop()
```

```python
win = Tk()
win.geometry("800x300")
win.option_add("*Font", "맑은고딕25")

#업종별 검색 라벨
indt_title = Label(win)
indt_title.config(text="< 업종별 검색 >")
indt_title.place(relx=0.4, rely=0.3)

#업종 입력창
def delete1(event):
    indt.delete(0, len(indt.get()))

def delete2(event):
    district.delete(0, len(district.get()))
    
indt = Entry(win, width=40)
indt.insert(0, "상권 분석을 원하는 업종명을 입력하세요.")
indt.bind("<Button-1>", delete1)
indt.place(relx=0.18,rely=0.45)

#거주 구 입력창
district = Entry(win, width=40)
district.insert(1, "현재 거주하고 있는 행정구를 입력하세요.")
district.bind("<Button-1>", delete2)
district.place(relx=0.18, rely=0.55)

#다음 화면으로 넘어가기
def search_result():
    global ind
    ind = indt.get()
    global district
    district = district.get()
    #for wg in win.place_slaves():
        #wg.destroy()
    result_page()

#검색 버튼 생성
searchbt = Button(win)
searchbt.config(width=5, height=2)
searchbt.config(text='검색')
searchbt.config(command = search_result)
searchbt.place(relx=0.75, rely=0.44)

#업종별 검색 결과
def result_page():
    
    rt_pg = Tk()
    rt_pg.geometry("800x600")
    rt_pg.option_add("*Font", "맑은고딕25")
    
    #결과 페이지 제목
    result_title = Label(rt_pg)
    result_title.config(text="[%s] 상권 분석 결과"%ind)
    result_title.grid(column=0, row=0, pady=40, columnspan=2, padx=50)
    
    #결과 페이지 top3, low3 설명 라벨
    rp_lab1 = Label(rt_pg)
    rp_lab1.config(text="[%s] 밀집 상권 top3"%ind)
    rp_lab1.grid(row=1, column=0, padx=40, pady=20)
    
    rp_lab2 = Label(rt_pg)
    rp_lab2.config(text="%s 내 [%s] 점포수 0개 상권"%(district,ind))
    rp_lab2.grid(row=1, column=1, pady=20)
    
    #top3 데이터 불러오기
    ind_lat = latest_df[latest_df['서비스_업종_코드_명'] == ind].sort_values('유사_업종_점포_수', axis=0, ascending=False)
    ind_lat_t = ind_lat.iloc[0:3,:]
    top3_name = ind_lat_t['상권_코드_명'].tolist()
    top3_stnum = ind_lat_t['유사_업종_점포_수'].tolist()
    top3_closer = ind_lat_t['폐업_률'].tolist()
    
    #top3 상권명 출력하기
    top_rst = produce_buttons(rt_pg, top3_name, 0)
    top_rst.init_widgets()

    #업종 없는 상권 데이터 불러오기
    ind_lat_l = ind_lat[ind_lat['상권_코드_명_구'] == district][ind_lat['유사_업종_점포_수'] == 0]
    none_name = ind_lat_l['상권_코드_명'].tolist()
    
    #없종 없는 상권명 출력하기
    none_rst = produce_buttons(rt_pg, none_name, 1)
    none_rst.init_widgets()


            
win.mainloop()
```
