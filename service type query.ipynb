```python
st = store.loc[:,'서비스_업종_코드':'서비스_업종_코드_명']
service_type = pd.DataFrame(st)
service_type.loc[:,'count'] = 1
stcount = service_type.groupby(['서비스_업종_코드_명', '서비스_업종_코드']).count()
stcount.reset_index(level=['서비스_업종_코드_명','서비스_업종_코드'], inplace=True)
stcount
```

```python
#서비스업종코드 개수 = 서비스업종명 개수 = 100개
#입력: 서비스업종명
#출력 1: 해당 서비스업종이 밀집된 상권코드명(top5) + 유사업종 점포 수 + 해당 업종 폐업률 + (추가데이터)생활인구 남/여 & 연령별 그래프
#출력 2: 해당 서비스업종이 부족한 상권코드명(top5) + 유사업종 점포 수 + 해당 업종 폐업률 + (추가데이터)생활인구 남/여 & 연령별 그래프

#연도별 구분을 어떻게 할지?! - 연도, 분기, 서비스업종 별로 groupby 후 출력에서 연도별 시계열 인사이트 제시?
# -> 인풋 받으면 그거에서 새로 데이터프레임 만들어야 할 듯
```

```python
#입력 값 예시
ex = '주유소'
ex_code = stcount.loc[stcount['서비스_업종_코드_명'] == ex].iloc[0,1]
ex_code
```

```python
#입력 값 예시 -> 해당 업종의 2021년 2분기 기준 점포 수 top5
ex_df = store.loc[store['서비스_업종_코드_명'] == ex].loc[store['기준_년_코드'] == '2021_2']
ex_df = ex_df.sort_values('유사_업종_점포_수', axis=0, ascending=False)
ex_top5 = ex_df.iloc[0:5,:]
ex_top5
```

```python
#입력 값 예시 -> 해당 업종의 2021년 2분기 기준 점포 수 low5
ex_low5 = ex_df.iloc[-100:-1,:]
ex_low5

#문제 -> 해당 지역에 점포 수가 0인 상권이 너무 많음
#사용자의 근처로 출력해주면 좋은데 상권 코드가 너무 다양해서 근처를 알 수 없을 듯
```

```python
#top5에 대해 점포수의 시계열 변화 나타내기
ex_top5_list = ex_top5.iloc[:, 6]
ex_top5_list.tolist()

#문제 -> 기준년코드와 기준분기코드를 하나로 합쳐야 시계열 분석이 편할 것 같음 -> 전처리 과정을 다시 손 대야 할 듯...
#엑셀에서 전처리로 바꾸고 옴 히힛

#시각화부터 시작하면 됨
```
