```python
# 지역별 검색
def dist():
    global district
    global indt
    dist = Toplevel()
    dist.geometry("800x300")
    dist.option_add("*Font", "맑은고딕 15")
    dist.title("서울시 상권 분석 서비스 - 지역별 검색")
    
    dist_bg = PhotoImage(file = "bg2.png")
    dist_l = Label(master = dist, image = dist_bg)
    dist_l.place(x = -2,y = -2)
    
#     indt_title = Label(dist)
#     indt_title.config(text="< 행정 구역별 검색 >")
#     indt_title.place(relx=0.37, rely=0.3)
    
    def delete1(event):
        district.delete(0, len(district.get()))
    
    #입력창 생성
    district = Entry(dist, width=40)
    district.insert(0, "상권 분석을 원하는 행정구를 입력하세요.")
    district.bind("<Button-1>", delete1)
    district.place(relx=0.18,rely=0.5)  
    
    def search_result():
        global district
        district = district.get()
        result_page()
    
    #검색 버튼 생성
    searchbt = Button(dist)
    searchbt.config(width=4, height=2)
    searchbt.config(text='검색')
    searchbt.config(command = search_result)
    searchbt.place(relx=0.75, rely=0.44)
    
    dist.mainloop()
    
#업종별 검색 결과
def result_page():
    global top3_name
    global low3_name
    
    rt_pg = Toplevel()
    rt_pg.geometry("700x400")
    rt_pg.option_add("*Font", "맑은고딕 15")
    rt_pg.title("서울시 상권 분석 서비스 - 지역별 검색 결과")
    
#     drp_bg = PhotoImage(file = "bg5.png")
#     drp_l = Label(master = rt_pg, image = drp_bg)
#     drp_l.place(x = -2,y = -2)
    
    #결과 페이지 제목
    result_title = Label(rt_pg)
    result_title.config(text="<%s 상권 분석 결과>"%district)
    result_title.grid(column=0, row=0, pady=40, columnspan=2, padx=50)
    
    #결과 페이지 top3, low3 설명 라벨
    rp_lab1 = Label(rt_pg)
    rp_lab1.config(text="%s 밀집 상권 top3"%district)
    rp_lab1.grid(row=1, column=0, padx=60, pady=20)
    
    rp_lab2 = Label(rt_pg)
    rp_lab2.config(text="%s 밀집 상권 low3"%district)
    rp_lab2.grid(row=1, column=1, padx=60, pady=20)
       
    #top3 데이터 불러오기
    rank = latest_df[latest_df['상권_코드_명_구']==district].groupby('서비스_업종_코드_명', as_index = True)['유사_업종_점포_수'].sum()
    df_ind= rank.sort_values(ascending = False).to_frame('count').reset_index().head(3)
    df_ind_2= rank.sort_values(ascending = True).to_frame('count').reset_index().head(3) 
    top3_name = df_ind['서비스_업종_코드_명'].tolist()
    low3_name = df_ind_2['서비스_업종_코드_명'].tolist()
    top3_count = df_ind['count'].tolist()
    low3_count = df_ind_2['count'].tolist()
    top3_list =[str(top3_name[0])+' : '+str(top3_count[0])+'개', str(top3_name[1])+' : '+str(top3_count[1])+'개', str(top3_name[2])+' : '+str(top3_count[2])+'개']
    list
    low3_list =[str(low3_name[0])+' : '+str(low3_count[0])+'개', str(low3_name[1])+' : '+str(low3_count[1])+'개', str(low3_name[2])+' : '+str(low3_count[2])+'개']
    list
    
    #top3 상권명 출력하기    
    button = Button(rt_pg,text = str(top3_name[0])+' : '+str(top3_count[0])+'개', command = clik_btn0)
    button.grid(column = 0, row = 2,padx = 40, pady = 20)
    button = Button(rt_pg,text = str(top3_name[1])+' : '+str(top3_count[1])+'개', command = clik_btn1)
    button.grid(column = 0, row = 3,padx = 40, pady = 20)
    button = Button(rt_pg,text = str(top3_name[2])+' : '+str(top3_count[2])+'개', command = clik_btn2)
    button.grid(column = 0, row = 4,padx = 40, pady = 20)
    
    
    #low3 상권명 출력하기    
    button = Button(rt_pg,text = str(low3_name[0])+' : '+str(low3_count[0])+'개', command = clik_btn3)
    button.grid(column = 1, row = 2,padx = 40, pady = 20)
    button = Button(rt_pg,text = str(low3_name[1])+' : '+str(low3_count[1])+'개', command = clik_btn4)
    button.grid(column = 1, row = 3,padx = 40, pady = 20)
    button = Button(rt_pg,text = str(low3_name[2])+' : '+str(low3_count[2])+'개', command = clik_btn5)
    button.grid(column = 1, row = 4,padx = 40, pady = 20)
    
def clik_btn0():
    rt_pg2 = Tk()
    rt_pg2.geometry("700x400")
    rt_pg2.option_add("*Font", "맑은고딕 15")
    rt_pg2.title("지역별 생활인구 분석")
    
#     drp2_bg = PhotoImage(file = "bg5.png")
#     drp2_l = Label(master = rt_pg2, image = drp2_bg)
#     drp2_l.place(x = -2,y = -2)
    
    #결과 페이지 제목
    result_title = Label(rt_pg2)
    result_title.config(text="<%s 업종 매출 분석 결과>"%district)
    result_title.grid(column=0, row=0, pady=40, columnspan=2, padx=50)
    
    # 매출정보 불러오기
    df3 = df2[df2['서비스_업종_코드_명']==top3_name[0]][['분기당_매출_금액','분기당_매출_건수']].aggregate([np.mean,max,min]).reset_index()       
    rp_lab1 = Label(rt_pg2)
    rp_lab1.config(text='%s의 분기당 평균 매출 금액은 %.3f원입니다.'%(top3_name[0],df3['분기당_매출_금액'].tolist()[0]))
    rp_lab1.grid(row=1, column=0, padx=40, pady=20)
    
    rp_lab2 = Label(rt_pg2)
    rp_lab2.config(text='%s의 분기당 평균 매출 건수는 %.3f건입니다.'%(top3_name[0],df3['분기당_매출_건수'].tolist()[0]))
    rp_lab2.grid(row=2, column=0, padx=40, pady=20)
    
    button = Button(rt_pg2,text = '성별에 따른 매출 분포', command = clik_btn1_1)
    button.grid(column = 0, row = 3,padx = 40, pady = 20)
    button = Button(rt_pg2,text = '요일에 따른 매출 분포', command = clik_btn1_2)
    button.grid(column = 0, row = 4,padx = 40, pady = 20)
    
        
    rt_pg2.mainloop()

def clik_btn1():
    rt_pg2 = Toplevel()
    rt_pg2.geometry("700x400")
    rt_pg2.option_add("*Font", "맑은고딕 15")
    rt_pg2.title("지역별 생활인구 분석")
    
#     drp2_bg = PhotoImage(file = "bg5.png")
#     drp2_l = Label(master = rt_pg2, image = drp2_bg)
#     drp2_l.place(x = -2,y = -2)
    
    #결과 페이지 제목
    result_title = Label(rt_pg2)
    result_title.config(text="<%s 업종 매출 분석 결과>"%district)
    result_title.grid(column=0, row=0, pady=40, columnspan=2, padx=50)
    
    # 매출정보 불러오기
    df3 = df2[df2['서비스_업종_코드_명']==top3_name[1]][['분기당_매출_금액','분기당_매출_건수']].aggregate([np.mean,max,min]).reset_index()       
    rp_lab1 = Label(rt_pg2)
    rp_lab1.config(text='%s의 분기당 평균 매출 금액은 %.3f원입니다.'%(top3_name[1],df3['분기당_매출_금액'].tolist()[0]))
    rp_lab1.grid(row=1, column=0, padx=40, pady=20)
    
    rp_lab2 = Label(rt_pg2)
    rp_lab2.config(text='%s의 분기당 평균 매출 건수는 %.3f건입니다.'%(top3_name[1],df3['분기당_매출_건수'].tolist()[0]))
    rp_lab2.grid(row=2, column=0, padx=40, pady=20)
    
    button = tk.Button(rt_pg2,text = '성별에 따른 매출 분포', command = clik_btn2_1)
    button.grid(column = 0, row = 3,padx = 40, pady = 20)
    button = tk.Button(rt_pg2,text = '요일에 따른 매출 분포', command = clik_btn2_2)
    button.grid(column = 0, row = 4,padx = 40, pady = 20)
    
    rt_pg2.mainloop()

def clik_btn2():
    rt_pg2 = Toplevel()
    rt_pg2.geometry("700x400")
    rt_pg2.option_add("*Font", "맑은고딕 15")
    rt_pg2.title("지역별 생활인구 분석")
    
#     drp2_bg = PhotoImage(file = "bg5.png")
#     drp2_l = Label(master = rt_pg2, image = drp2_bg)
#     drp2_l.place(x = -2,y = -2)
    
    #결과 페이지 제목
    result_title = Label(rt_pg2)
    result_title.config(text="<%s 업종 매출 분석 결과>"%district)
    result_title.grid(column=0, row=0, pady=40, columnspan=2, padx=50)
    
    # 매출정보 불러오기
    df3 = df2[df2['서비스_업종_코드_명']==top3_name[2]][['분기당_매출_금액','분기당_매출_건수']].aggregate([np.mean,max,min]).reset_index()       
    rp_lab1 = Label(rt_pg2)
    rp_lab1.config(text='%s의 분기당 평균 매출 금액은 %.3f원입니다.'%(top3_name[2],df3['분기당_매출_금액'].tolist()[0]))
    rp_lab1.grid(row=1, column=0, padx=40, pady=20)
    
    rp_lab2 = Label(rt_pg2)
    rp_lab2.config(text='%s의 분기당 평균 매출 건수는 %.3f건입니다.'%(top3_name[2],df3['분기당_매출_건수'].tolist()[0]))
    rp_lab2.grid(row=2, column=0, padx=40, pady=20)
    
    button = tk.Button(rt_pg2,text = '성별에 따른 매출 분포', command = clik_btn3_1)
    button.grid(column = 0, row = 3,padx = 40, pady = 20)
    button = tk.Button(rt_pg2,text = '요일에 따른 매출 분포', command = clik_btn3_2)
    button.grid(column = 0, row = 4,padx = 40, pady = 20)
    rt_pg2.mainloop()
    
def clik_btn3():
    global close
    rt_pg3 = Tk()
    rt_pg3.geometry("400x400")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 폐업률")
    
    #폐업률 그래프 그리기
    sns.set_palette("bright")
    close = store[store['상권_코드_명_구']==district].groupby(['서비스_업종_코드_명','기준_년_코드'], as_index = True)['폐업_률'].mean()
    figure = plt.Figure(figsize=(4,3), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    close[low3_name[0]].plot(legend = True,ax=ax)
    df3 = store[store['서비스_업종_코드_명']==low3_name[0]]['폐업_률'].aggregate([np.mean,max,min]).reset_index()
    ax.set_title(str(low3_name[0])+'의'+' '+'평균 폐업률은'+' '+str(round(df3['폐업_률'].tolist()[0],2))+'입니다')
        
    rt_pg3.mainloop()
        
def clik_btn4():
    global close
    rt_pg3 = Tk()
    rt_pg3.geometry("400x400")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 폐업률")
    
    #폐업률 그래프 그리기
    sns.set_palette("bright")
    close = store[store['상권_코드_명_구']==district].groupby(['서비스_업종_코드_명','기준_년_코드'], as_index = True)['폐업_률'].mean()
    figure = plt.Figure(figsize=(4,3), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    close[low3_name[1]].plot(legend = True,ax=ax)
    df3 = store[store['서비스_업종_코드_명']==low3_name[1]]['폐업_률'].aggregate([np.mean,max,min]).reset_index()
    ax.set_title(str(low3_name[1])+'의'+' '+'평균 폐업률은'+' '+str(round(df3['폐업_률'].tolist()[0],2))+'입니다')
        
    rt_pg3.mainloop()

def clik_btn5():
    global close
    rt_pg3 = Tk()
    rt_pg3.geometry("400x400")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 폐업률")
    
    #폐업률 그래프 그리기
    sns.set_palette("bright")
    close = store[store['상권_코드_명_구']==district].groupby(['서비스_업종_코드_명','기준_년_코드'], as_index = True)['폐업_률'].mean()
    figure = plt.Figure(figsize=(4,3), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    close[low3_name[2]].plot(legend = True,ax=ax)
    df3 = store[store['서비스_업종_코드_명']==low3_name[2]]['폐업_률'].aggregate([np.mean,max,min]).reset_index()
    ax.set_title(str(low3_name[2])+'의'+' '+'평균 폐업률은'+' '+str(round(df3['폐업_률'].tolist()[0],2))+'입니다')
        
    rt_pg3.mainloop()

def clik_btn1_1():
    rt_pg3 = Tk()
    rt_pg3.geometry("600x600")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 성별 매출")
    
    sns.set_palette("bright")
    figure = plt.Figure(figsize=(6,6), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    sex = df2[df2['서비스_업종_코드_명']==top3_name[0]][['남성_매출_비율','여성_매출_비율']]
    sex.plot(kind = 'box',legend = True,ax=ax)
    ax.set_title(str(top3_name[0])+'의 성별 매출금액 입니다')
        
    rt_pg3.mainloop() 
    
def clik_btn1_2():
    rt_pg3 = Tk()
    rt_pg3.geometry("600x600")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 요일별 매출")
    
    figure = plt.Figure(figsize=(6,6), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    day = df2[df2['서비스_업종_코드_명']== top3_name[0]][['월요일','화요일','수요일','목요일','금요일','토요일','일요일']]
    day2 = day[['월요일','화요일','수요일','목요일','금요일','토요일','일요일']].describe().loc['mean']
    day2.plot(kind = 'bar',legend = True,ax=ax,color="darkgreen")
    ax.set_title(str(top3_name[0])+'의 요일별 매출금액 입니다')
        
    rt_pg3.mainloop()    
    
def clik_btn2_1():
    rt_pg3 = Tk()
    rt_pg3.geometry("600x600")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 성별 매출")

    sns.set_palette("bright")
    figure = plt.Figure(figsize=(6,6), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    sex = df2[df2['서비스_업종_코드_명']==top3_name[1]][['남성_매출_비율','여성_매출_비율']]
    sex.plot(kind = 'box',legend = True,ax=ax)
    ax.set_title(str(top3_name[1])+'의 성별 매출금액 입니다')
        
    rt_pg3.mainloop() 
    
def clik_btn2_2():
    rt_pg3 = Tk()
    rt_pg3.geometry("600x600")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 요일별 매출")
    
    figure = plt.Figure(figsize=(6,6), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    day = df2[df2['서비스_업종_코드_명']== top3_name[1]][['월요일','화요일','수요일','목요일','금요일','토요일','일요일']]
    day2 = day[['월요일','화요일','수요일','목요일','금요일','토요일','일요일']].describe().loc['mean']
    day2.plot(kind = 'bar',legend = True,ax=ax,color="darkgreen")
    ax.set_title(str(top3_name[1])+'의 요일별 매출금액 입니다')
        
    rt_pg3.mainloop()    
    
def clik_btn3_1():
    rt_pg3 = Tk()
    rt_pg3.geometry("600x600")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 성별 매출")
    
    sns.set_palette("bright")
    figure = plt.Figure(figsize=(6,6), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    sex = df2[df2['서비스_업종_코드_명']==top3_name[2]][['남성_매출_비율','여성_매출_비율']]
    sex.plot(kind = 'box',legend = True,ax=ax)
    ax.set_title(str(top3_name[2])+'의 성별 매출금액 입니다')
        
    rt_pg3.mainloop() 
    
def clik_btn3_2():
    rt_pg3 = Tk()
    rt_pg3.geometry("600x600")
    rt_pg3.option_add("*Font", "맑은고딕 15")
    rt_pg3.title("지역별 생활인구 분석 결과 - 요일별 매출")
    
    figure = plt.Figure(figsize=(6,6), dpi =100)
    ax = figure.add_subplot(111)
    line = FigureCanvasTkAgg(figure, rt_pg3)
    line.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    day = df2[df2['서비스_업종_코드_명']== top3_name[2]][['월요일','화요일','수요일','목요일','금요일','토요일','일요일']]
    day2 = day[['월요일','화요일','수요일','목요일','금요일','토요일','일요일']].describe().loc['mean']
    day2.plot(kind = 'bar',legend = True,ax=ax,color="darkgreen")
    ax.set_title(str(top3_name[2])+'의 요일별 매출금액 입니다')
        
    rt_pg3.mainloop()       
```
